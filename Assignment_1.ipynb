```python
from sklearn.preprocessing import StandardScaler 
```

```python
sc=StandardScaler()

sepal_width = df['sepal_width'].values.reshape(-1,1)
tranformed_sepal_width=sc.fit_transform(sepal_width)
tranformed_sepal_width
tranformed_sepal_width
```

```python
z_score=((3.0-df['sepal_width'].mean())/df['sepal_width'].std())
z_score 
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()

sepal_width = df['sepal_width'].values.reshape(-1, 1)
tranformed_sepal_width=sc.fit_transform(sepal_width)
tranformed_sepal_width
```

```python
z_score=((3.0-df['sepal_width'].min())/(df['sepal_width'].max()-df['sepal_width'].min()))
z_score
```
